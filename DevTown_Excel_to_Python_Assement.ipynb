# KPI Analysis

Q3. Calculate the following KPIs:

```python
# Total Sales
df['TotalAmount'].sum()
```

```python
# Total Profit
df["Profit"] = df["TotalAmount"] - ((df["Quantity"] * df["UnitPrice"]) - df['Discount'] - df['Tax'] - df['ShippingCost'])
df['Profit'].sum()
```

```python
# Total Orders
df['OrderID'].count()
```

```python
# Average Sales
df['TotalAmount'].mean()
```

```python
#	Average Profit
df['Profit'].mean()
```

```python
#	Maximum Sales
df['TotalAmount'].max()
```

```python
#	Minimum Sales
df['TotalAmount'].min()
```

# Time-based Analysis

Q12. Calculate the monthly sales trend using the Order Date column.

```python
# Convert the date column to datetime
df["OrderDate"] = pd.to_datetime(df["OrderDate"])

# Extract month number (1–12)
df["Month"] = df["OrderDate"].dt.month

df.groupby('Month')['TotalAmount'].sum()
```

# Visualization

Q15. Create a Bar Chart showing Sales by State.

```python
df.groupby('State')['TotalAmount'].sum().plot(kind = 'bar')
plt.title('Sales by State')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```

Q16. Create a Pie Chart showing Category-wise Sales.

```python
df.groupby('Category')['TotalAmount'].sum().plot(kind = 'pie')
plt.title('Sales by Region')
plt.show()
```

Q17. Create a Line Chart showing Monthly Sales Trend.

```python
df.groupby('Month')['TotalAmount'].sum().plot(kind = 'line')
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

Q18. Create a Horizontal Bar Chart showing the Top 10 Products by Sales.

```python
df.groupby('ProductName')['TotalAmount'].sum().sort_values(ascending=False).head(10).plot(kind = 'barh')
plt.title('Sales by State')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.show()
```
